```python
dt=movies.merge(credit,on="title")      #merging two dataframe on the basis of TITLE
```

```python
dt=dt[["genres","movie_id","title","popularity","cast","crew","overview","keywords"]] #extracting the data from main dataframe and save it into the same
dt.shape
```

```python
dt.isnull().sum()
```

```python
dt.dropna(inplace=True)  #dropping the null data
```

```python
dt.iloc[0].genres  #extracting data from genres only
```

```python
# since here we get list of dictionary and the list is in string format,so changing it into integer using (ast.literal_eval)
import ast

def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i["name"])
  return l
```

```python
dt["genres"]=dt["genres"].apply(convert)
```

```python
dt["keywords"]=dt["keywords"].apply(convert)
```

```python
dt.sample(3)
```

```python
dt["cast"][0]
```

```python
# here we have to take only 3 values of dictionary 
def convert3(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter <3:
      l.append(i["name"])
    counter+=1
   
  return l
```

```python
dt["cast"]=dt["cast"].apply(convert3)
```

```python
dt.sample(3)
```

```python

dt["crew"][0]
```

```python
# now we are taking those value only where job type is director only
def fetch_director(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i["job"]=="Director":
      l.append(i["name"])
      break
  return l
```

```python
dt["crew"]=dt["crew"].apply(fetch_director)
```

```python
dt.sample(3)
```

```python
dt.sample(1)
```

```python
# here we join/eliminate the space between the names as same name can be appeared so avoiding that we are doing that

dt["genres"]=dt["genres"].apply(lambda x:[i.replace(" ","")for i in x])
dt["keywords"]=dt["keywords"].apply(lambda x:[i.replace(" ","")for i in x])
dt["cast"]=dt["cast"].apply(lambda x:[i.replace(" ","")for i in x])
dt["crew"]=dt["crew"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
dt.sample(3)
```

```python
dt["overview"]=dt["overview"].apply(lambda x:x.split()) #converting the overvier column into list using split function
```

```python
dt["tags"]=dt["genres"]+dt["keywords"]+dt["crew"]+dt["cast"]+dt["overview"]    #cocatenate/merging the required features in tags 
```

```python
dt.sample(3)
```

```python
new_dt=dt[["movie_id","title","tags"]]   #making new dataframe and extracting the required feature
```

```python
new_dt["tags"]=new_dt["tags"].apply(lambda x:" ".join(x))   #replacing the commas with space using apply function 
```

```python
new_dt.sample(3)
```

```python
new_dt["tags"][0]
```

```python
new_dt["tags"]=new_dt["tags"].apply(lambda x:x.lower())    #converting the tags feature in lowercase
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv=CountVectorizer (max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(new_dt["tags"]).toarray()
```

```python
cv.get_feature_names()  #it gives the top 5000 words which are used frequently
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)   #gives similarity of each movies w.r.t others
```
